# MatplotLib

```python
# Library for creating static, animated, and interactive visualizations
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Used to directly show graphs in jupyter
%matplotlib inline
```

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## Plotting using matplotlib

plt.scatter(x,y,c='orange')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('graph.png')
```

```python
## Plt plot
y=x*x

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
## Creating Subplots
y=x*x

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'k*--')
```

```python
x=np.arange(-10,11,0.1)

y=(x+3)*(x-3)
plt.title('Quadratic Equation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y,'b')
```

```python
## Plot sine waves

pi=np.pi
x=np.arange(0,4*pi,0.1)

plt.subplot(2,1,1)
y=np.sin(x)
plt.title('Sine Curve')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y,'g--')

plt.subplot(2,1,2)
y=np.cos(x)
plt.title('Cosine Curve')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y,'r--')
```

```python
## Simultaneous Plots
pi=np.pi
x=np.arange(0,4*pi,0.1)

y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,'g--')
plt.plot(x,y2,'r--')

plt.title('Sinosoidal Curve')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Bar Graph

```python
## Bar Plot

x1=[2,8,10]
y1=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x1,y1,color='g')
plt.bar(x2,y2,color='b')

plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Histogram

```python
a=np.array([15,20,32,18,26,25,17,31,29,28,15,21,23,25,24])
plt.hist(a,bins=10)

plt.title('Histogram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Box Plot

```python
## Useful to represent percentiles
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

## Pie Chart

```python
# Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) # explode first slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)

plt.axis('equal')
```
