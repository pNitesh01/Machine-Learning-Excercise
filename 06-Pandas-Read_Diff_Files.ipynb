# Pandas Reading Different Files

## Json

```python
import pandas as pd
import numpy as np

data = '[ { "id": "A001", "name": "Tom", "math": 60, "physics": 66, "chemistry": 61 }, { "id": "A002", "name": "James", "math": 89, "physics": 76, "chemistry": 51 }, { "id": "A003", "name": "Jenny", "math": 79, "physics": 90, "chemistry": 78 }]'

df = pd.read_json(data)
df
```

```python
## Use orient='record', otherwise json with indexes is created

df.to_json('student.json',orient="records")
df
```

## HTML

```python
url = 'https://en.wikipedia.org/wiki/Minnesota'

dfHtml = pd.read_html(url, match='Election results from statewide races')

dfHtml[0]
```

```python
dfHtml[0].to_html("election-data.html")
```
