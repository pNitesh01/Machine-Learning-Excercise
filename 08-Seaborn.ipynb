# Seaborn

# Distribution plots
* joinplot
* pairplot
* displot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
## Correlation uses numerals only
df.corr()
```

## Correlation with Heatmap

```python
## Shows a 2D correlation matrix between two discrete dimensions, using colored cells to represent data from usually a monochromatic scale.
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

## Pair Plot

```python
sns.pairplot(df)
```

```python
## Pair plot based on a parameter
sns.pairplot(df,hue='sex')
```

```python
df['sex'].value_counts()
```

## Dis Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical plots
* countplot
* barplot
* boxplot
* violinplot

```python
df.head()
```

## Count Plot

```python
## Count Plot
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

## Bar Plot

```python
## Bar Plot
sns.barplot(x='day',y='total_bill',data=df)
```

## Box Plot

```python
## Useful for percentiles
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
## Categorize data based on other category

sns.boxplot(x='day',y='total_bill',data=df,hue='sex',palette='Set3')
```

## Violin Plot

```python
## Combination of box plot and kernel density function

sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```
